#  For loops

## 1. Regular loop

```python
list = []
for i in range(11):
    list.append(i**3)
```

```python
list
```

## 2. List loop

```python
list = ['dog', 'cat', 'mouse', 'horse']
for item in list:
    print(item)
```

## 3. Enumeration loop

```python
list = ['dog', 'cat', 'mouse', 'horse']
for i, item in enumerate(list):
    print('Index {} contains {}'.format(i, item))
```

## 4. List comprehension loop

```python
list = [i**3 for i in range(11)]
```

```python
list
```

## 5. Double for loop

```python
for i in range(3):
    for j in range(3):
        print('i = {} j = {}'.format(i,j))
```
